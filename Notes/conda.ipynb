Here are some useful conda commands and notes:

### Conda Environment Management

1. **Create a new environment:**
    ```sh
    conda create --name <env_name> python=<version>
    ```
    Example:
    ```sh
    conda create --name py010 python=3.12
    ```

2. **Activate an environment:**
    ```sh
    conda activate <env_name>
    ```
    Example:
    ```sh
    conda activate py010
    ```

3. **Deactivate the current environment:**
    ```sh
    conda deactivate
    ```

4. **List all environments:**
    ```sh
    conda env list
    ```

5. **Remove an environment:**
    ```sh
    conda remove --name <env_name> --all
    ```
    Example:
    ```sh
    conda remove --name py010 --all
    ```

### Package Management

1. **Install a package:**
    ```sh
    conda install <package_name>
    ```
    Example:
    ```sh
    conda install ipykernel
    ```

2. **Remove a package:**
    ```sh
    conda remove <package_name>
    ```
    Example:
    ```sh
    conda remove py010
    ```

3. **Update a package:**
    ```sh
    conda update <package_name>
    ```

4. **List installed packages:**
    ```sh
    conda list
    ```

### Jupyter Notebook Integration

1. **Install `ipykernel` in the environment:**
    ```sh
    conda install ipykernel
    ```

2. **Create a Jupyter kernel for the environment:**
    ```sh
    python -m ipykernel install --user --name <env_name> --display-name "Python (<env_name>)"
    ```
    Example:
    ```sh
    python -m ipykernel install --user --name py010 --display-name "Python (py010)"
    ```

### Configuration

1. **Set conda to always yes:**
    ```sh
    conda config --set always_yes yes
    ```

2. **Add a new channel:**
    ```sh
    conda config --add channels <channel_name>
    ```

3. **Remove a channel:**
    ```sh
    conda config --remove channels <channel_name>
    ```

4. **List all configurations:**
    ```sh
    conda config --show
    ```

These commands should help you manage your conda environments and packages effectively.
