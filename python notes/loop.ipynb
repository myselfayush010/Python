### Code Examples:



```python


# 1. Basic For Loop
for i in range(5):
    print(i)  # Prints 0 to 4

# 2. For Loop with List
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

# 3. While Loop
count = 0
while count < 5:
    print(count)
    count += 1

# 4. Break Statement
for i in range(10):
    if i == 5:
        break
    print(i)

# 5. Continue Statement
for i in range(5):
    if i == 2:
        continue
    print(i)

# 6. Nested Loops
for i in range(3):
    for j in range(3):
        print(f"({i},{j})")

# 7. Loop with Enumerate
for index, value in enumerate(fruits):
    print(f"Index: {index}, Value: {value}")

# 8. List Comprehension
squares = [x**2 for x in range(5)]

# 9. Dictionary Comprehension
cube_dict = {x: x**3 for x in range(5)}

# 10. While Loop with else
x = 0
while x < 3:
    print(x)
    x += 1
else:
    print("Done!")

# 11. For Loop with else
for i in range(3):
    print(i)
else:
    print("Loop completed!")

# 12. Infinite Loop with Break
while True:
    response = input("Enter 'quit' to exit: ")
    if response == 'quit':
        break
```



### Common Use Cases:



```python


# Iterating through multiple lists
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f"{name} is {age} years old")

# File iteration
with open('file.txt', 'r') as file:
    for line in file:
        print(line.strip())

# Dictionary iteration
person = {'name': 'John', 'age': 30, 'city': 'New York'}
for key, value in person.items():
    print(f"{key}: {value}")

# Loop with step value
for i in range(0, 10, 2):  # Step of 2
    print(i)  # Prints 0,2,4,6,8

# Filter with list comprehension
numbers = [1, 2, 3, 4, 5, 6]
evens = [x for x in numbers if x % 2 == 0]
```
